```python
from sklearn.preprocessing import OneHotEncoder

def get_ohe(df, col):
    ohe = OneHotEncoder(drop='first', handle_unknown='error', sparse_output=False, dtype='int')
    ohe.fit(df[[col]])
    temp_df = pd.DataFrame(data=ohe.transform(df[[col]]), columns=ohe.get_feature_names_out())
    # If you have a newer version, replace with columns=ohe.get_feature_names_out()
    df.drop(columns=[col], axis=1, inplace=True)
    df = pd.concat([df.reset_index(drop=True), temp_df], axis=1)
    return df
```

```python
main_data = get_ohe(main_data, 'Gender')
main_data = get_ohe(main_data, 'Customer Type')
main_data = get_ohe(main_data, 'Type of Travel')
main_data.head(20)
print(main_data.columns)
```

```python
penalty_data = pd.read_csv('C:/Users/udayv/Downloads/Invistico_Airline.csv')
penalty_data.dropna(inplace=True)
penalty_data.isna().sum()
```

```python
# Filter the dataset for 'Inflight entertainment' equal to 0
penalty_data = penalty_data[penalty_data['Inflight entertainment'] == 0]

# Define features and target variable
X = penalty_data.drop(columns=['satisfaction', 'Class', 'Inflight entertainment'])
y = penalty_data['Class']

# Perform one-hot encoding for categorical columns
X_encoded = pd.get_dummies(X, drop_first=True)

# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=50, stratify=y)

# Scale the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Check the number of rows in the filtered dataset
num_rows = len(penalty_data)
print("Number of rows after filtering:", num_rows)
```

```python

log_reg = LogisticRegression(fit_intercept = True,multi_class='ovr', solver='lbfgs', penalty=None)

# Fit the model on the training data
log_reg.fit(X_train_scaled, y_train)

log_reg.score(X_train_scaled, y_train)
log_reg.score(X_test_scaled,y_test)
```

```python
# Initialize the logistic regression model with L1 penalty
log_reg_l1 = LogisticRegression(penalty='l1', solver='liblinear', C=0.1, multi_class='ovr', random_state=0)

# Fit the model on the training data
log_reg_l1.fit(X_train_scaled, y_train)

# Calculate the test & train accuracy
train_accuracy_l1 = log_reg_l1.score(X_train_scaled,y_train)
test_accuracy_l1 = log_reg_l1.score(X_test_scaled, y_test)
print("Test Accuracy with L1 Penalty:", test_accuracy_l1)
print("Train Accuracy with L1 Penalty:",train_accuracy_l1)
# This is the coefficient Beta_1, ..., Beta_7
log_reg_l1.coef_

# This is the coefficient Beta_0
log_reg_l1.intercept_

```

```python
# Get the coefficients and intercept of the model
coefficients = log_reg_l1.coef_
intercept = log_reg_l1.intercept_

# Get the feature names
feature_names = X.columns

# Create a DataFrame to display the coefficients for each feature
coefficients_df = pd.DataFrame(coefficients, columns=feature_names)

# Display the coefficients
print("Coefficients:")
print(coefficients_df)

# Identify features with (nearly) zero coefficients
zero_coefficients = coefficients_df.columns[coefficients_df.iloc[0] == 0]
print("\nFeatures with (nearly) zero coefficients:", zero_coefficients)
```

```python
# Initialize the logistic regression model with L2 penalty
log_reg_l2 = LogisticRegression(penalty='l2', solver='lbfgs', C=0.1, multi_class='ovr', random_state=0)

# Fit the model on the training data
log_reg_l2.fit(X_train_scaled, y_train)

# Calculate the test accuracy
test_accuracy_l2 = log_reg_l2.score(X_test_scaled, y_test)
print("Test Accuracy with L2 Penalty:", test_accuracy_l2)

# Calculate the train accuracy
train_accuracy_l2 = log_reg_l2.score(X_train_scaled, y_train)
print("Train Accuracy with L2 Penalty:", train_accuracy_l2)
```
