Applying OneHotEncoder on the catogorical independent variables i.e., Gender, Customer Type, Type of Travel.

```python
from sklearn.preprocessing import OneHotEncoder

def get_ohe(df, col):
    ohe = OneHotEncoder(drop='first', handle_unknown='error', sparse_output=False, dtype='int')
    ohe.fit(df[[col]])
    temp_df = pd.DataFrame(data=ohe.transform(df[[col]]), columns=ohe.get_feature_names_out())
    df.drop(columns=[col], axis=1, inplace=True)
    df = pd.concat([df.reset_index(drop=True), temp_df], axis=1)
    return df
```

```python
subset_data = get_ohe(subset_data, 'Gender')
subset_data = get_ohe(subset_data, 'Customer Type')
subset_data = get_ohe(subset_data, 'Type of Travel')
subset_data.head()
```

```python
print(f"Number of Independent variables: {subset_data.shape[1] - 1}")
```

we have 21 independent variables in the dataset after one-hot encoding.

Splitting the data into test and train by stratify 'Class' and with random_state as 50.

```python
subset_data = get_ohe(subset_data, 'Gender')
subset_data = get_ohe(subset_data, 'Customer Type')
subset_data = get_ohe(subset_data, 'Type of Travel')
subset_data.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(subset_data.drop(columns = ['Class','Inflight entertainment']), subset_data['Class'], test_size=0.20, stratify=subset_data['Class'],random_state=50)

X_train
X_test
y_train
y_test
first_train_index = X_train.index[0]
print(f"The start index for train dataset is {first_train_index}")
first_test_index = X_test.index[0]
print(f"The start index for test dataset is {first_test_index}")
```

```python
model = LogisticRegression(fit_intercept = True, solver='lbfgs', multi_class = 'ovr', penalty = None)
# model = LogisticRegression(fit_intercept = True, solver='lbfgs', multi_class = 'ovr', penalty = 'none', max_iter = 1000)
# model = LogisticRegression(fit_intercept = True, solver='lbfgs', multi_class = 'multinomial', penalty = 'none', max_iter = 1000)
# model = LogisticRegression(fit_intercept = True, solver='newton-cg', multi_class = 'multinomial', penalty = 'none', max_iter = 10000)

# While using multiclass case do multi_class = 'ovr' or 'auto'; can also try other solvers
# While doing regularization, use penalty = 'l2' and also C = 10.0 (need to try other values too)

model.fit(X_train, y_train) 

# The following gives the mean accuracy on the given data and labels
train_score = model.score(X_train, y_train) 
test_score = model.score(X_test, y_test)
train_score_percentage = "{:.2f}".format(train_score * 100)
test_score_percentage = "{:.2f}".format(test_score * 100)
print(f"Accuracy for train dataset is: {train_score_percentage} %")
print(f"Accuracy value for test dataset is: {test_score_percentage} %")

# This is the coefficient Beta_1, ..., Beta_7
model.coef_

# This is the coefficient Beta_0
model.intercept_
```
