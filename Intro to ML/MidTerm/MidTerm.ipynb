```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=['Demand']), df['Demand'], test_size=0.20, random_state = 9729)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_scaled = pd.DataFrame(sc.fit_transform(X_train), columns = X_train.columns, index = X_train.index)
X_test_scaled = pd.DataFrame(sc.transform(X_test), columns = X_test.columns, index = X_test.index)
```

```python
model = LinearRegression(fit_intercept = True)
```

```python
model.fit(X_train_scaled, y_train) 

# The following gives the mean accuracy on the given data and labels
model.score(X_train_scaled, y_train) 

# This is the coefficient Beta_1, ..., Beta_7
model.coef_

# This is the coefficient Beta_0
model.intercept_
```

e) Name: Saisri Vishwanath Email: savishwa@syr.edu SUID: 980432838

Answer: The R-square and intercept for training data are 0.56 and 143.51

```python
test_output = pd.DataFrame(model.predict(X_test_scaled), index = X_test_scaled.index, columns = ['pred_Demand'])
test_output.head()
```

```python
test_output = test_output.merge(y_test, left_index = True, right_index = True)
test_output.head()
print('Percentage of correct predictions is ')
print(model.score(X_test_scaled, y_test))
```

```python
mean_absolute_error = abs(test_output['pred_Demand'] - test_output['Demand']).mean()
print('Mean absolute error is ')
print(mean_absolute_error)
```

```python
average_demand_test = y_test.mean()
```

```python
fraction_mae = mean_absolute_error / average_demand_test
print("Fraction of MAE to Average Demand:", fraction_mae)
```
