**1. Read data from CSV file using pandas**

```python
from google.colab import drive
drive.mount("/drive", force_remount=True)
import csv
import pandas as pd
trueArticle = pd.read_csv('/drive/My Drive/Colab Notebooks/IST 664/2023/True.csv')

#Extract only the text field from the data"
true = trueArticle["text"]
print(true)
len(true), type(true[0])
```

**2.PRE-PROCESSING**


Merge the data from all articles to one single string

```python
trueText = ''
for i in range(len(true)):
    trueText += true[i]

print(trueText[0:1000])
len(trueText)
```

There are 51042676 characters in all the articles together

Tokenization

To separate data into individual words I am using "punkt" tokenizer which separates by spaces and special characters(non-alphabetical characters)

```python
import nltk
nltk.download('punkt')
trueTokens=nltk.word_tokenize(trueText)
print(len(trueTokens), trueTokens[0:1000])
```

Lower-case

To maintain uniformity and consistency among the tokens format for better understanding and processing I am using islower() which makes all token to lower case leters.

```python
trueWords = [w.lower() for w in trueTokens]
print(trueWords[0:1000])
```

Removing non-words(non-alphabetical words) or Getting Total words
as non-alphabetical characters don't add any weightage to the content

```python
resivedtrueWords = [w for w in trueWords if w.isalpha()]
print(len(resivedtrueWords))
```

There 8033628 words in all the articles together

Removing stop-words

Removing words which doesn't add any weightage to the content of the sentence

```python
nltk.download('stopwords')
nltk_stops = nltk.corpus.stopwords.words('english')
truecontentw = [w for w in resivedtrueWords if w not in nltk_stops] #content words
print(len(truecontentw))
```

There are 4690368 words after removing stop works from the articles

Lemmatization to reduce words to their root word("lemma") despite spellings and pronounciations

```python
nltk.download('wordnet')
nltk.download('omw-1.4')
wnl = nltk.WordNetLemmatizer()
type(wnl)
```

```python
truelemma =[wnl.lemmatize(t) for t in trueTokens]

#percentage reduction achieved with lemmatization
round( 100 - (len(set(truelemma))/len(set(trueTokens)) * 100), 1)
```

**3. DATA ANALYSIS**

List of top50 stop words and content words by their frequencies

```python
from nltk import FreqDist
truestopw = [w for w in trueWords if w in nltk_stops] #stop words

truestopwfdist = FreqDist(truestopw)
truecontentwfdist = FreqDist(truecontentw)

toptruestopw = truestopwfdist.most_common(50)
toptruecontentw = truecontentwfdist.most_common(50)

print("top 50 stop words")
for pair in toptruestopw:
    print (pair)

print("top 50 content words")
for pair in toptruecontentw:
    print (pair)
```

From the results, it's clear that after removing stop words and non-alphabetical characters, the frequency ditribution of 50 most common words is the words that hold meaning to the sentence/corpus and also the frequency distribution of stop words(verbose) is more than frequery distribution of content words

List of top50 bigrams by frequencies

```python
import re
from nltk.collocations import *
from nltk.collocations import BigramAssocMeasures
from nltk.collocations import BigramCollocationFinder

def alpha_filter(w):
  # pattern to match word of non-alphabetical characters
  pattern = re.compile('^[^a-z]+$')
  if (pattern.match(w)):
    return True
  else:
    return False

bigram_measures = nltk.collocations.BigramAssocMeasures()
finder = BigramCollocationFinder.from_words(trueWords)
scored1 = finder.score_ngrams(bigram_measures.raw_freq)
finder.apply_word_filter(alpha_filter) #filter non-aplhabetical characters
scored2 = finder.score_ngrams(bigram_measures.raw_freq)
finder.apply_word_filter(lambda w: w in nltk_stops) #filter stop words
scored3 = finder.score_ngrams(bigram_measures.raw_freq)

print("Top 50 Bigrams by frequency")
for i in range(50):
    print (scored1[i])

print("Top 50 Bigrams by frequemcy after removing punctuations")
for i in range(50):
    print (scored2[i])

print("Top 50 Bigrams by frequency after removing punctuations and stop words")
for i in range(50):
    print (scored3[i])
```

From result, we can say that frequency distribution of bigram with punctuations and stops words is more than frequery distribution of bigram after removing

Top 50 bigrams by PMI Measure

```python
bigram_measures = nltk.collocations.BigramAssocMeasures()
finder4 = BigramCollocationFinder.from_words(trueWords)
finder4.apply_freq_filter(5)
truepmi = finder4.score_ngrams(bigram_measures.pmi)

print("Top 50 PMI Measure")
for i in range(50):
    print (truepmi[i])
```

Bigrams from PMI are different from bigrams by frequency and bigrams by PMI have very high PMI score(in the range of 20)

Top 50 Adjective words

I am getting the top50 adjective words by using pos_tag

```python
from nltk import pos_tag
nltk.download('averaged_perceptron_tagger')
pos_tags = pos_tag(trueTokens)
adjtags = [ w for w in pos_tags if w[1] == 'JJ']
adjdist = FreqDist(adjtags)
topadjdist = adjdist.most_common(50)
for i in topadjdist:
  print(i)
```

**Counts**

```python
trueArticleTokens = [None]*len(true)
for i in range(len(true)):
    trueArticleTokens[i]=nltk.word_tokenize(true[i]);
print(len(trueArticleTokens))
print(trueArticleTokens[0],len(trueArticleTokens[0]))
```

```python
import re
def alpha_filter(w):
  # pattern to match word of non-alphabetical characters
  pattern = re.compile('^[^a-z]+$')
  if (pattern.match(w)):
    return True
  else:
    return False

def punctuations(w):
  # pattern to match word of non-alphabetical characters
  pattern = re.compile('^[^a-z]+$')
  if (not(w.isnumeric()) and pattern.match(w)):
    return True
  else:
    return False

def caps_letter(w):
  if (w !="I" and w != 'i'):
      if(w.isupper()):
          return True
      else:
          return False
  else:
    return False


def exclamation(w):
    if( '!' in w):
        return True
    else:
        return False

truetotalwords = []*len(trueArticleTokens)
truecontentwords = [] * len(trueArticleTokens)
totalcaps  = [] * len(trueArticleTokens)
totalexclaim  = [] * len(trueArticleTokens)
totalpunct = [] * len(trueArticleTokens)
for i in range(len(trueArticleTokens)):
    currtruetotalwords = 0
    currtruecontentwords = 0
    currtruecapswords = 0
    currtrueexclaimwords = 0
    currtruepunctwords = 0
    for j in  range(len(trueArticleTokens[i])):
        w = trueArticleTokens[i][j]
        if (not(punctuations(w))):
            currtruetotalwords+=1
        if (not(punctuations(w)) and w not in nltk_stops):
            currtruecontentwords+=1
        if (caps_letter(w)):
            currtruecapswords +=1
        if (exclamation(w)):
            currtrueexclaimwords +=1
        if( punctuations(w)):
            currtruepunctwords +=1
    truetotalwords.append(currtruetotalwords)
    truecontentwords.append(currtruecontentwords)
    totalcaps.append(currtruecapswords)
    totalexclaim.append(currtrueexclaimwords)
    totalpunct.append(currtruepunctwords)

```

```python
print("Total number of words")
for i in range(len(truetotalwords)):
    print (truetotalwords[i])

print("Total number of content words")
for i in range(len(truecontentwords)):
    print (truecontentwords[i])

print("Total number of capitalized words")
for i in range(len(totalcaps)):
    print (totalcaps[i])

print("Total number of exclaimation words")
for i in range(len(totalexclaim)):
    print (totalexclaim[i])

print("Total number of punctuations")
for i in range(len(totalpunct)):
    print (totalpunct[i])
```

Saving the counts to CSV file

```python
csv_file = '/drive/My Drive/Colab Notebooks/IST 664/2023/True.csv'
column_name1 = "TotalWords"
column_name2 = "TotalcontentWords"
column_name3 = "TotalcapitalizedWords"
column_name4 = "TotalexclaimationWords"
column_name5 = "TotalPunctuations"
trueArticle[column_name1] = truetotalwords
trueArticle[column_name2] = truecontentwords
trueArticle[column_name3] = totalcaps
trueArticle[column_name4] = totalexclaim
trueArticle[column_name5] = totalpunct

# Save the DataFrame to the CSV file
trueArticle.to_csv(csv_file, index=False)
```
