**1. Read data from CSV file using pandas**

```python
from google.colab import drive
drive.mount("/drive", force_remount=True)
import csv
import pandas as pd
fakeArticle = pd.read_csv('/drive/My Drive/Colab Notebooks/IST 664/2023/Fake.csv')

#Extract onlt the text field from the data"
fake = fakeArticle["text"]
print(fake)
len(fake), type(fake[0])
```

**2.PRE-PROCESSING**

Merge the data from all articles to one single string

```python
fakeText = ''
for i in range(len(fake)):
    fakeText += fake[i]

print(fakeText[0:1000])
len(fakeText)
```

There are 59815411 characters in all the articles together

Tokenization

To separate data into individual words I am using "punkt" tokenizer which separates by spaces and special characters(non-alphabetical characters)

```python
import nltk
nltk.download('punkt')
fakeTokens=nltk.word_tokenize(fakeText)
print(len(fakeTokens), fakeTokens[0:1000])
```

Lower-case

To maintain uniformity and consistency among the tokens format for better understanding and processing I am using islower() which makes all token to lower case leters.

```python
fakeWords = [w.lower() for w in fakeTokens]
print(fakeWords[0:1000])
```

Removing non-words(non-alphabetical words) or Getting Total words
as non-alphabetical characters don't add any weightage to the content

```python
resivedfakeWords = [w for w in fakeWords if w.isalpha()]
print(len(resivedfakeWords))
```

There are 9579621 together in all the articles together

Removing stop-words


Removing words which doesn't add any weightage to the content of the sentence

```python
nltk.download('stopwords')
nltk_stops = nltk.corpus.stopwords.words('english')
fakecontentw = [w for w in resivedfakeWords if w not in nltk_stops] #content words
print(len(fakecontentw))
```

There are 5168292 words after removing stop words in the articles

Lemmatization to reduce words to their root word("lemma") despite spellings and pronounciations

```python
nltk.download('wordnet')
nltk.download('omw-1.4')
wnl = nltk.WordNetLemmatizer()
type(wnl)
```

```python
fakelemma =[wnl.lemmatize(t) for t in fakeTokens]

#percentage reduction achieved with lemmatization
round( 100 - (len(set(fakelemma))/len(set(fakeTokens)) * 100), 1)
```

**3. DATA ANALYSIS**

List of top50 stop words and content words by their frequencies

```python
from nltk import FreqDist
fakestopw = [w for w in fakeWords if w in nltk_stops] #stop words

fakestopwfdist = FreqDist(fakestopw)
fakecontentwfdist = FreqDist(fakecontentw)

topfakestopw = fakestopwfdist.most_common(50)
topfakecontentw = fakecontentwfdist.most_common(50)

print("top 50 stop words")
for pair in topfakestopw:
    print (pair)

print("top 50 content words")
for pair in topfakecontentw:
    print (pair)
```

From the results, it's clear that after removing stop words and non-alphabetical characters, the frequency ditribution of 50 most common words is the words that hold meaning to the sentence/corpus and also the frequency distribution of stop words(verbose) is more than frequery distribution of content words

List of top50 bigrams by frequencies

```python
import re
from nltk.collocations import *
from nltk.collocations import BigramAssocMeasures
from nltk.collocations import BigramCollocationFinder

def alpha_filter(w):
  # pattern to match word of non-alphabetical characters
  pattern = re.compile('^[^a-z]+$')
  if (pattern.match(w)):
    return True
  else:
    return False

bigram_measures = nltk.collocations.BigramAssocMeasures()
finder = BigramCollocationFinder.from_words(fakeWords)
scored1 = finder.score_ngrams(bigram_measures.raw_freq)
finder.apply_word_filter(alpha_filter) #filter non-alphabetical characters
scored2 = finder.score_ngrams(bigram_measures.raw_freq)
finder.apply_word_filter(lambda w: w in nltk_stops) #filter stop words
scored3 = finder.score_ngrams(bigram_measures.raw_freq)

print("Top 50 Bigrams by frequency")
for i in range(50):
    print (scored1[i])

print("Top 50 Bigrams by frequemcy after removing punctuations")
for i in range(50):
    print (scored2[i])

print("Top 50 Bigrams by frequency after removing punctuations and stop words")
for i in range(50):
    print (scored3[i])
```

From result, we can say that frequency distribution of bigram with punctuations and stops words is more than frequery distribution of bigram after removing

Top 50 bigrams by PMI measure

```python
bigram_measures = nltk.collocations.BigramAssocMeasures()
finder4 = BigramCollocationFinder.from_words(fakeWords)
finder4.apply_freq_filter(5)
fakepmi = finder4.score_ngrams(bigram_measures.pmi)

print("Top 50 PMI Measure")
for i in range(50):
    print (fakepmi[i])
```

**Counts**

```python
fakeArticleTokens = [None]*len(fake)
for i in range(len(fake)):
    fakeArticleTokens[i]=nltk.word_tokenize(fake[i]);
print(len(fakeArticleTokens))
print(fakeArticleTokens[0],len(fakeArticleTokens[0]))
```

```python
import re
def alpha_filter(w):
  # pattern to match word of non-alphabetical characters
  pattern = re.compile('^[^a-z]+$')
  if (pattern.match(w)):
    return True
  else:
    return False

def punctuations(w):
  # pattern to match word of non-alphabetical characters
  pattern = re.compile('^[^a-z]+$')
  if (not(w.isnumeric()) and pattern.match(w)):
    return True
  else:
    return False

def caps_letter(w):
  if (w !="I" and w != 'i'):
      if(w.isupper()):
          return True
      else:
          return False
  else:
    return False


def exclamation(w):
    if( '!' in w):
        return True
    else:
        return False

faketotalwords = []*len(fakeArticleTokens)
fakecontentwords = [] * len(fakeArticleTokens)
faketotalcaps  = [] * len(fakeArticleTokens)
faketotalexclaim  = [] * len(fakeArticleTokens)
faketotalpunct = [] * len(fakeArticleTokens)
for i in range(len(fakeArticleTokens)):
    currfaketotalwords = 0
    currfakecontentwords = 0
    currfakecapswords = 0
    currfakeexclaimwords = 0
    currfakepunctwords = 0
    for j in  range(len(fakeArticleTokens[i])):
        w = fakeArticleTokens[i][j]
        if (not(punctuations(w))):
            currfaketotalwords+=1
        if (not(punctuations(w)) and w not in nltk_stops):
            currfakecontentwords+=1
        if (caps_letter(w)):
            currfakecapswords +=1
        if (exclamation(w)):
            currfakeexclaimwords +=1
        if( punctuations(w)):
            currfakepunctwords +=1
    faketotalwords.append(currfaketotalwords)
    fakecontentwords.append(currfakecontentwords)
    faketotalcaps.append(currfakecapswords)
    faketotalexclaim.append(currfakeexclaimwords)
    faketotalpunct.append(currfakepunctwords)
```

```python
print("Total number of words")
for i in range(len(faketotalwords)):
    print (faketotalwords[i])

print("Total number of content words")
for i in range(len(fakecontentwords)):
    print (fakecontentwords[i])

print("Total number of capitalized words")
for i in range(len(faketotalcaps)):
    print (faketotalcaps[i])

print("Total number of exclaimation words")
for i in range(len(faketotalexclaim)):
    print (faketotalexclaim[i])

print("Total number of punctuations")
for i in range(len(faketotalpunct)):
    print (faketotalpunct[i])
```

Saving the counts to CSV file

```python
csv_file = '/drive/My Drive/Colab Notebooks/IST 664/2023/Fake.csv'
column_name1 = "TotalWords"
column_name2 = "TotalcontentWords"
column_name3 = "TotalcapitalizedWords"
column_name4 = "TotalexclaimationWords"
column_name5 = "TotalPunctuations"
fakeArticle[column_name1] = faketotalwords
fakeArticle[column_name2] = fakecontentwords
fakeArticle[column_name3] = faketotalcaps
fakeArticle[column_name4] = faketotalexclaim
fakeArticle[column_name5] = faketotalpunct

# Save the DataFrame to the CSV file
fakeArticle.to_csv(csv_file, index=False)
```
